```python
import windtools
```

# Create a synthetic power curve from a capacity and rotor diameter

```python
pc1 = windtools.SyntheticPowerCurve( capacity=4200, rotordiam=136)
```

```python
pc1
```

# Change the cut-out wind speed

```python
pc2 = windtools.SyntheticPowerCurve( capacity=4200, rotordiam=136, cutout=30)
```

```python
pc2.
```

# Use a known specific capacity to make the power curve

```python
pc3 = windtools.SyntheticPowerCurve( specificCapacity=350 )
```

```python
pc3
```

# Plot all power curves together to compare

```python
import matplotlib.pyplot as plt
ax = plt.subplot(111)

pc1.plot( ax=ax, color="red", linestyle="--", linewidth=3, label="With Cap. and RD.")
pc2.plot( ax=ax, color="blue", linestyle="-", linewidth=1.5, label="With higher cutout")
pc3.plot( ax=ax, color="green", linestyle="-", linewidth=1.5, label="With specific Cap.")

plt.legend()

plt.show()
```
